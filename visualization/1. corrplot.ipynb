## Corrplot
* Corrplot으로 correlation test 결과를 시각화하자
---

```r
library('corrplot')
```

```r
df <- as.matrix(read.csv('./Hippo signalling pathway (human).csv', row.names = 1))
pval <- as.matrix(read.csv('./Hippo signalling pathway (human)_pval.csv', row.names = 1))
```

```r
corrplot(df, p.mat = pval,    #correlation value와 p value 불러오기
         insig = "label_sig", sig.level = c(.001, .01, .05), pch.cex = .9, pch.col = "black",    #p value 시각화 (star mark)
         is.corr = FALSE, method = 'color', tl.col = "black",    #negative value만 있기 때문에 is.corr = FALSE. x/y text: black
         cl.ratio = 0.2, cl.cex = 0.8)    #legend 크기 조절
```
